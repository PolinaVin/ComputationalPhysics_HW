# Интерполяция

```python
def Ln(x, k, Xk):
    y = 1
    for i in range (len(Xk)) :
        if i != k:
            y*=((x - Xk[i])/ (Xk[k]-Xk[i]))
    return y

def F(Xnew, Xs, Ys):
    return sum (Ys[i]*Ln(Xnew, i, Xs) for i in range(len(Xs)))
```

## Интерполяция равномерной сеткой

```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
```

```python
def Draw(dots, begin=-1, end=1):
    x_data = np.linspace (-1, 1, dots)
    y_data = [1/(25*x**2 + 1) for x in x_data]
    x_list = np.linspace (-1, 1, 700)
    f = [F(i, x_data, y_data) for i in x_list]
    plt.plot (x_data, y_data, "o", x_list, f, '-')
    plt.show()
```

```python
Draw(10)#По 10 точкам
```

```python
Draw(20)#По 20 точкам
```

```python
Draw(100)#По 100 точкам
```

## Интерполяция узлами Чебышева

```python
def Draw_by_knots(knots):
    x_data = [(np.cos(np.pi*(2*n - 1)/(2*knots))) for n in range(1, knots+1)]
    y_data = [1/(25*i**2 + 1) for i in x_data]
    x_list = np.linspace (-1, 1, 500)
    f=[(F(i, x_data, y_data)) for i in x_list]
    plt.plot (x_data, y_data, "o", x_list, f, '--')
    plt.show()
```

```python
Draw_by_knots(10)#По 10 узлам
```

```python
Draw_by_knots(20)#По 20 узлам
```

```python
Draw_by_knots(100)#По 100 узлам
```
