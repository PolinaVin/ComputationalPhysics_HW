```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
def f(x,u, const):
    return -const[0]**2*x-const[1]*u+const[2]*x**3
```

```python
print('input const')
const=list(map(float, input().split(',')))
print('input baseline', 'x_0=', sep='\n', end='')
x_0 = float(input())
print('u_0=', end='')
u_0 = float(input())
h=np.linspace(0.0, 10*2*pi/const[0], 10**(6))
const.append(h[1]-h[0])
```

# Euler Method

```python
def Euler(x_i, u_i, const):
    u_new = u_i + const[3] * f(x_i, u_i, const)
    return u_new

x=[x_0]
u=[u_0]

for t_i in h[1:]:
    u.append(Euler(x[-1], u[-1], const))
    x.append(x[-1] + const[3] * u[-2])
plt.plot(h, x, '-')
```

# Runge_Kutta Method

```python
def Runge_Kutta(x_i, u_i, const):
    u_new = u_i + const[3] * f(0.5 * const[3] + x_i, u_i + 0.5 * const[3] * f(x_i , u_i, const), const)
    return u_new

x= [x_0]
u= [u_0]

for t_i in h[1:]:
    u.append(Runge_Kutta(x[-1], u[-1], const))
    x.append(x[-1] + const[3] * u[-2])
plt.plot(h, x, '-', label='$y = Runge_Kutta')
plt.legend()
```
